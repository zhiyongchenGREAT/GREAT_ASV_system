# M-MFCC Api

## M-MFCC端点检测方法的接口文档

```python
# librosa
import librosa
import librosa.display
import IPython.display

# python_speech_features
import python_speech_features
from python_speech_features import sigproc

# numpy
import numpy
import numpy as np

# plt
import matplotlib.pyplot as plt

# conv
from scipy import signal
```

```python
# 预处理 分帧
def frames(signal,samplerate=16000,winlen=0.025,winstep=0.01,preemph=0.97,
          winfunc=lambda x:numpy.ones((x,))):
    signal = sigproc.preemphasis(signal,preemph)
    return sigproc.framesig(signal, winlen*samplerate, winstep*samplerate, winfunc)
```

```python
# 音频读取
y, sr = librosa.load("ori_00000.wav", sr=16000)

time = [i/sr for i in range(y.shape[0])]
```

```python
# 时域图
plt.plot(time, y)
plt.ylabel("Amplitude", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.xlabel("Time(s)", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.title("WaveForm", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.grid()
IPython.display.Audio(data=y, rate=sr)
```

```python
# 计算 mfcc_n
mfcc_n = python_speech_features.mfcc(y, sr)
```

```python
# MFCC图
librosa.display.specshow(mfcc_n.T)
```

```python
# 计算前20帧的均值向量,基本要求前 0.2s不能有语音（先验条件）
average_vector = mfcc_n[0].copy()
for i in range(1, 20):
    average_vector += mfcc_n[i]
average_vector /= 20
```

```python
# 计算M_n
frame = frames(y, sr)
frame_abs = np.abs(frame)
M_n = np.mean(frame_abs, axis=1)
```

```python
# 计算MFCC（语音帧与噪声帧的）余弦相似度的距离矩阵
# d(m_noise, m_i)
distance = []
a = 0.97 # 动态噪声更新的调节因子
for i in range(len(mfcc_n)):
    # 更新噪声
    m_noise_i = a * average_vector + (1 - a) * mfcc_n[i]
    # 计算余弦距离
    distance_i = np.abs((np.linalg.norm(m_noise_i)*(np.linalg.norm(mfcc_n[i]))/np.dot(m_noise_i, mfcc_n[i])))
    distance.append(distance_i)
```

```python
# MFCC
plt.plot(distance)
plt.ylim([0,10])
plt.ylabel("Distance", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.xlabel("sample", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.title("Frame Distance with Noise", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
```

```python
# 卷积平滑distance
distance_conv = signal.convolve(distance, np.array(
    [1, 1, 1, 1, 1, 1, 1]
))[3:-3]
```

```python
plt.plot(distance_conv)

plt.ylabel("Distance", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.xlabel("sample", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.title("Frame Distance with Noise Add convolve", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })

```

```python
# M-MFCC均衡
# 均衡因子p
p = 0.02 / np.average(M_n) if np.average(M_n) > 0.02 else 1

M_MFCC = (1 - p)*(distance_conv/max(distance_conv)) + p*(M_n/max(M_n))
```

```python
plt.plot(M_MFCC, label="Mn-MFCC")
plt.legend(loc = 'best')
plt.ylim([0,1])
plt.ylabel("Mn-MFCC Coefficient", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.xlabel("sample", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
plt.title("Mn-MFCC", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
```

```python
# 切割起止点
# 阈值
# 用户可调节阈值调节因子
k = 1
Threshold = np.average(M_MFCC) * k
out = np.zeros_like(M_MFCC)
for i in range(10, len(M_MFCC) - 10):
    out[i] = 1 if np.average(M_MFCC[i-10:i+10]) > Threshold else 0
```

```python
plt.plot([Threshold]*len(M_MFCC), label="Threshold="+"%.2f" % Threshold)
plt.plot(M_MFCC, label="Mn-MFCC")
plt.plot(out, label="VAD")
plt.legend(loc = 'upper right')
plt.xlabel("sample", {'family': 'Times New Roman',
                         'weight': 'normal',
                         'size': 16,
                         })
```

```python
# 计算每帧对应时间
res = []
start_flag = 0
end_flag = 0
start = 0.
end = 0.
for i in range(1, len(out)):
    if start_flag == 0 and end_flag == 0:
        if out[i] == 1 and out[i-1] == 0:
            # 起始点
            start = i/100
            start_flag = 1
    elif start_flag == 1 and end_flag == 0:
        if  out[i] == 0 and out[i-1] == 1:
            # 终止点
            end = i/100
            end_flag = 1
    elif start_flag == 1 and end_flag == 1:
        start_flag = 0
        end_flag = 0
        res.append({
            "start": start,
            "end": end
        })

for i in res:
    print(i)



```
